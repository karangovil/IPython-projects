## Loading datasets

We will be using a [gene expression](http://en.wikipedia.org/wiki/Gene_expression) data set obtained from a [microarray](http://en.wikipedia.org/wiki/DNA_microarray) experiement [Read more about the specific experiment here](http://www.ncbi.nlm.nih.gov/geo/query/acc.cgi?acc=GSE5859).  There are two data sets we will use:  

1. The gene expression intensities where the rows represent the features on the microarray (e.g. genes) and the columns represent the different microarray samples.  

2. A table that contains the information about each of the samples (columns in the gene expression data set) such as the sex, the age, the treatment status, the date the samples were processed.  Each row represents one sample. 

```python
exprs = pd.read_csv('exprs_GSE5859.csv', index_col=0)

sampleinfo = pd.read_csv('sampleinfo_GSE5859.csv')
```

```python
exprs.head()
```

```python
sampleinfo.head()
```

We can see that the order of columns in `exprs` does not match the `filename` order in `sampleinfo`. Let us change it so it matches.

```python
a = list(sampleinfo.filename)
b = list(exprs.columns)
matchIndex = [b.index(x) for x in a]
exprs = exprs[matchIndex]

#check if all columns in exprs are in same order as filenames in sampleinfo
(exprs.columns == sampleinfo.filename).all()
```

In order to find the number of days since October 31, 2002, we need to separate the months and years in `sampleinfo`. Let us now extract the year and month as integers from the `sampleinfo` table.

```python
sampleinfo["date"] = pd.to_datetime(sampleinfo.date)
sampleinfo["month"] = map(lambda x:x.month, sampleinfo.date)
sampleinfo["year"] = map(lambda x:x.year, sampleinfo.date)

oct31 = pd.datetime(2002,10,31,0,0)

sampleinfo["elapsedInDays"] = map(lambda x:(x-oct31).days, sampleinfo.date)
sampleinfo.head()
```

## Exploratory data analysis and SVD

We will now try to determine if the variable `elapsedInDays` has a significant effect on the variability in the data or if it is just the ethnicity. 

```python
#here we subset only the CEU ethnicity data
sampleinfoCEU = sampleinfo[sampleinfo.ethnicity == 'CEU']
exprsCEU = exprs[sampleinfoCEU.filename]
#check to make sure the order of the columns in exprsCEU matches the rows in sampleinfoCEU dataframe
(exprsCEU.columns == sampleinfoCEU.filename).all()
```

Next we normalize the gene expression intensity in `exprsCEU` by subtracting the means.

```python
exprsCEU_normalized = exprsCEU.apply(lambda x:(x - exprsCEU.mean(axis=1)), axis=0)
exprsCEU_normalized.head()
```

Using the normalized data, we can perform the Singular Value decomposition using the linalg functions of SciPy and get the first principal axis PC1

```python
U,s,Vh = linalg.svd(exprsCEU_normalized.values)
V = Vh.T
```

Let us now create a scatterplot of PC1 vs `elapsedInDays` to see if there is any effect of number of days on the variability in data

```python
plt.scatter(sampleinfoCEU.elapsedInDays, V[:,0])
plt.title('Relationship between the PC1 and the date the samples were processed')
plt.xlabel('Date sample was processed (Number of days since Oct 31, 2012)')
plt.ylabel('PC1')
plt.xlim(0,1000)
```

It appears that something interesting happens around day 160, so let us zoom in on that

```python
plt.scatter(sampleinfoCEU.elapsedInDays, V[:,0])
plt.title('Relationship between the PC1 and the date the samples were processed')
plt.xlabel('Date sample was processed (Number of days since Oct 31, 2012)')
plt.ylabel('PC1')
plt.xlim(0,160)
plt.axvline(x=100, color='r')
```

Thus we observe that there is a difference in the way samples were processed around day 100. Using exploratory data analysis and the SVD of the gene expression data matrix, we see the date the samples were processed does have a large effect on the variability seen in the data.
