## Monte Carlo simulations to estimate the percentage for Obama votes

### Single poll simulation

Now we will perform a Monte Carlo simulation to obtain the estimated percentage of Obama votes with a sample size $N$ where $N$ is the median sample size calculated above. Let $p$=0.53 be the percent of voters are voting for Obama.

```python
p = 0.53
B = 1000
# we do the MC simulation by 2 ways 
mcarlo1 = np.random.binomial(N,p,size=B) / N
mcarlo2 = map(lambda x: np.mean(np.random.binomial(1, p, size = N)), xrange(B))
plt.hist(mcarlo1)
plt.hist(mcarlo2)
```

This ditribution looks pretty normal so let us now check it using Q-Q plot

```python
import scipy.stats as stats
stats.probplot((mcarlo1 - np.mean(mcarlo1)) / np.std(mcarlo1, ddof=1), dist="norm", plot=plt)
plt.show()
```

Thus from the Q-Q plot and histogram, we can conclude that the distrbution is normal.

We know that if we observe $N$ Bernoulli random variables $(X_1, \ldots X_N)$, then 
$$ \mbox{E}(\bar{X}) = \frac{1}{N} \sum_{i=1}^N p = p$$
and 
$$\mbox{Var}(\bar{X})= \frac{1}{N^2} \sum_{i=1}^N p(1-p) = \frac{p(1-p)}{N}$$

In our example, we assume each $X_i$ is a Bernoulli distribution with $p$ = 0.53. Therefore, if $N$ = 1200, we can analytically calculate the standard deviation of $\bar{X}$ directly and compare to the standard error of the simulation.

```python
sim_std = np.std(mcarlo1, ddof=1)
theor_std = np.sqrt(0.53*0.47/N)
print "The analytic value of standard deviation is %s and the simulated value is %s" %(theor_std, sim_std)
```

### Multiple poll simulation

We found earlier that there was a total of $M = 18$ independent pollster polls with a median poll size of $N = 1200$ in November 2012. Within each iteration, we want to simulate percentage of Obama votes from $M$ polls and then compute the mean. We will do 1000 such simulations and plot the results.

```python
#function to do one simulation
def simulatePolls(p,N,M):
    return map(lambda x: np.mean(np.random.binomial(1,p,size=N)), xrange(M))

mcarlo3 = map(lambda x: np.mean(simulatePolls(p,N,M)), xrange(B))
plt.hist(mcarlo3)
plt.show()
stats.probplot((mcarlo3 - np.mean(mcarlo3)) / np.std(mcarlo3, ddof=1), dist="norm", plot = plt)
plt.show()
print "The standard deviation of 18 polls is %s which is %s times smaller than that of a single poll." \
    %(np.round(np.std(mcarlo3,ddof=1),5), np.round(np.std(mcarlo1,ddof=1) / np.std(mcarlo3,ddof=1),5))
```

Let us also compute standard deviations across the 18 polls and make sure its also normally distributed.

```python
mcarlo3_sd = map(lambda x: np.std(simulatePolls(p,N,M)), xrange(B))
plt.hist(mcarlo3_sd)
plt.show()
stats.probplot((mcarlo3_sd - np.mean(mcarlo3_sd)) / np.std(mcarlo3_sd, ddof=1), dist="norm", plot = plt)
plt.show()
```

### Standard deviation of $M$ polls in real 2012 pollster data

```python
pollster_sd = np.std(filtered["Obama"] / 100)
print "The standard deviation in the pollster data is %s which is %s times the simulated value" \
        %(pollster_sd, np.round(pollster_sd / np.mean(mcarlo3_sd),5))
```

In this problem we considered all the polls from pollsters in November 2012. After removing polls that came from the same pollster in Nov 2012, we did not see any pollster bias.  The standard deviation from the 2012 Presidential Election was smaller than what we expect by chance.  

# Is the average of polls better than just one poll?

Most undecided voters vote for one of the two candidates at the election. Therefore, the reported percentages underestimate the final value of both candidates. However, if we assume the undecided will split evenly, then the observed difference should be an unbiased estimate of the final difference. 

Let us make a plot of the differences for the week before the election (e.g. 5 days) where the days are on the x-axis and the differences are on the y-axis. We also show a horizontal line showing 3.9%: the difference between Obama and Romney on election day.

```python
election["Diff"] = (election.Obama / 100.0) - (election.Romney / 100.0)

last_day = max(election["Start Date"])
filtered = election[map(lambda x: (last_day - x).days <= 5, election["Start Date"])]
filtered = filtered.sort(columns="Start Date")
days = map(lambda x: (last_day - x).days, filtered["Start Date"])

color_map = {}
for i, p in enumerate(set(filtered.Pollster)):
    color_map[p] = np.random.rand();

plt.scatter(days, filtered.Diff, c = map(lambda x: color_map[x], filtered.Pollster),  s=60 )
plt.axhline(y=0.039, c = "gray")
plt.text(-0.5, 0.04, "Diff = 3.9% on day of election")
plt.axhline(y=np.mean(filtered.Diff), c = "red")
plt.text(-0.5, 0.0127, "Mean difference across pollsters")
plt.xlabel("Days")
plt.ylabel("Difference (Obama - Romney)")
plt.title("Plot of the difference between Obama and Romney colored by different pollsters in the last week")
```

Let us also make a plot showing the differences by pollster where the pollsters are on the x-axis and the differences on the y-axis.

```python
pollster_map = {}

polls = list(set(filtered.Pollster))

for i, p in enumerate(polls):
    pollster_map[p] = i

plt.scatter(map(lambda x: pollster_map[x],filtered.Pollster), filtered.Diff, \
            c = map(lambda x: color_map[x],filtered.Pollster),s=60)
plt.xticks(range(len(polls)), polls, rotation = 90)
plt.xlabel("Pollsters")
plt.ylabel("Difference (Obama - Romney)")
plt.title("Plot of the difference between Obama and Romney by different pollsters")
plt.show()
```

We can now compare the variability within each pollster (across a set of polls) to the variability between each pollster. From these two visualizations it is clear that the between pollster difference is larger.
